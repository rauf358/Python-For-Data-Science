Welcome to our Virtual Bank System project! This project utilizes the fundamentals of object-oriented programming (OOP) to create a simulated banking environment. Through this system, users can perform various banking operations like deposits, withdrawals, currency exchanges, payments, and checking account details. It's a straightforward and efficient way to experience virtual banking functionalities,incorporating robust exception handling to ensure smooth operation even in unexpected scenarios.


```python
exchange_rates = {
    'EUROS': 1.12,  # 1 Euro = 1.12 USD
    'British Pounds': 1.33,  # 1 British Pound = 1.33 USD
    'Canadian Dollars': 0.79,  # 1 Canadian Dollar = 0.79 USD
    'Australian Dollars': 0.71,  # 1 Australian Dollar = 0.71 USD
    'Japanese Yen': 0.0091  # 1 Japanese Yen = 0.0091 USD
}
```

```python
class Bank_Account:
    
    def __init__(self,name,acc_num):
        self.name= name
        self.acc_num=acc_num
        self.balance = {"Account number":acc_num,'Balance': 0 }
        
    def deposit(self,amount):
        self.balance['Balance'] = self.balance['Balance'] + amount
        print(f"Account#{self.balance['Account number']} the balance is ${self.balance['Balance']}\n")
        
    def Withdraw(self,tran):
        self.balance['Balance'] = self.balance['Balance'] - tran
        
        try:   
            if self.balance['Balance'] < 0 :
                self.balance['Balance'] = 0
                raise ValueError("No money present in Account")
            else:
                print(f"From Account No# {self.balance['Account number']} ${tran} is Withdrawn\n")
                print(f"\nAccount#{self.balance['Account number']} the balance remaining is ${self.balance['Balance']}\n")    
                
        except ValueError as e:
            print(f"Error : {e} \n")
    
    def currency_exchge(self,amount,exc,exc2):
        self.balance['Balance'] = self.balance['Balance'] - amount
        
        try:   
            if self.balance['Balance'] < 0 :
                self.balance['Balance'] = 0
                raise ValueError("No money present in Account\n")
            else:
                usd = amount/exc
                self.balance[exc2] = usd
                print(f"Converted Money {exc2} {usd}\n")
                print(f"\nAccount#{self.balance['Account number']} the balance remaining is ${self.balance['Balance']}\n")    

        except ValueError as e :
             print(f"Error : {e} \n")
    def payment(self,pl,p):
           
        try:   
            if self.balance['Balance'] < 0 :
                self.balance['Balance'] = 0
                raise ValueError("No money present in Account\n")
            else:
                self.balance['Balance'] = self.balance['Balance'] - p
                print(f"$\n{p} amount is transfered to {pl}")
                print(f"\nAccount#{self.balance['Account number']} the balance remaining is ${self.balance['Balance']}\n")    

        except ValueError as e :
             print(f"Error : {e} \n")
                
    def show_details(self):
        
        print('\n')
        
        for k,v in self.balance.items():
            print(f"{k} : {v}\n")
        
```

```python
def bank():    
    print("\t****Hello, Wellcome!,to Virtual Bank****")
    
    Name= input("\nWhats your name ? ")
    
    account_num = input(f"Enter Your Account Number {Name} \n")
    
    try:
        if len(account_num) !=11:
            raise ValueError("Invaild Account Number")
            
        elif not account_num.isdigit:
            raise ValueError("Invaild Input")
            
    except ValueError as e:
        print(f"Error : {e}")
        return
    
    person = Bank_Account(Name,account_num)
   
    commands = ['deposit','Withdraw','currency exchange','payment','show_details','exit']
    
    print("\nList of commands\n")
    
    while True:
        
        for i in range(0,len(commands)):
            print(f'{i+1}) {commands[i]}')
        action = int(input("\n What do you want to do ? "))
        
        if action == 1:
            try:
                am = int(input("Input Amount\n"))
                
            except ValueError:
                print("Invalid Input\n")
                continue
                
            person.deposit(am)
            
        if action == 2:
            try:
                tm = int(input("Input Amount to Withdrawn "))
                
            except ValueError:
                print("Invalid Input\n")
                continue
                
            person.Withdraw(tm)
            

        if action == 3:
            
            print("\nExchange rates\n")

            for k,v in exchange_rates.items():
                print(f"{k}:{v}\n")
                
            try:
                amt = int(input("Enter amount to transfer \n"))
                exc = input("Specify currency \n")
                
            except ValueError:
                print("Invalid Input\n")
                continue
                
            person.currency_exchge(amt,exchange_rates[exc],exc)
        
        if action ==4:
            pay_loc = input("\nPayment Address : ")
            pay = int(input("\nAmount Payment :$"))
            person.payment(pay_loc,pay)
        if action == 5:
            person.show_details()
        
        if action == 6:
            print("****Thanks For Using Our Website Please Come Again!!****")
            break
```

```python
bank()
```
