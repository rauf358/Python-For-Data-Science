The program utilizes the Kaggle API to fetch a dataset and applies regular expression filtering to extract specific data. It first authenticates the Kaggle API, downloads the dataset, and loads it into a pandas DataFrame. Then, it defines a regular expression pattern and filters the DataFrame based on a specified column using the pattern. Finally, the filtered data is displayed.

```python
import pandas as pd
```

```python
patient= pd.read_csv("Clinical Data_Discovery_Cohort.csv")
patient
```

```python
 pattern = r'[Pp]?T[0-9][PN][0-9]?[NM][0-9]?X?' # regular expression pattern
```

```python
patient_bol=patient['Stage'].str.contains(pattern)
patient_bol
```

```python
 patient_stage_PT=patient[patient_bol]
patient_stage_PT
```

```python
patient_Alive=patient['Dead or Alive'].str.contains("Alive")
patient_Alive
```

```python
patient[patient_Alive]
```
