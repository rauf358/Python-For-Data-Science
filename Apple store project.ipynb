In this project I have OOPS(Object Oriented Programming) concept to make web store of Apple items 

```python
 class Cart:
    def __init__(self,name):
        self.name=name
        self.items={ 'sku1':{'iPhone':0,'price':600},
                    'sku2':{'iMac':0,'price':400},
                    'sku3':{'Airpods':0,'price':200},
                    'sku4':{'Apple Watch':0,'price':900},
                    'sku5':{'iPod':0,'price':700},
                    'sku6':{'Apple Tv':0,'price':900},
                    'sku7':{'Macbook':0,'price':1200},
                   }
        self.cart_items={}
        self.price_sku = []
        self.total = 0 
    def add(self,sku):
        
        try:
            key =next(iter(self.items['sku'+ sku]))
            self.price_sku.append(sku)        
            if key not in self.cart_items:
                self.cart_items[key] = 1
            else :
                self.cart_items[key] += 1
            print(f'{key} is added to cart')
        except:
            print("Invalid Option")
        
    def remove(self,sku):
        try:
            self.cart_items.pop(sku)
            self.price_sku.remove(sku)        
            print(f'\n{sku} is removed from cart')
        except KeyError:
            print("The item you want to remove is not in the cart  \n")
        except:
            print("Please enter the item you want to remove \n")
        
    def prices(self):
        for i in self.price_sku:
            print(f'{next(iter(self.items["sku"+str(i)])) }: {self.items["sku"+str(i)]["price"]}')
            self.total =self.total + int(self.items["sku"+str(i)]["price"])
    def show_cart(self):
        if self.cart_items != {}: 
            for i,v in self.cart_items.items():
                print(f"{i} : {v}")
        else:
            print("Cart is Empty")
    def check_out(self):
        print(f"\nYou're sum total is ${self.total} dollers! ")
        print(f"thanks for Using Our Platform {self.name}\n\tcome again!!")
```

```python
def order():
    
    print("Wellcome to Apple Store")
    
    valid_commands = ['add','remove','prices','show cart','check out']
    
    customer_name=input("May I know you're name ? ")
    
    customer = Cart(customer_name)
    
    true=True
    
    while true:
        for i in range(len(valid_commands)):
            print(f'\n{i+1}: {valid_commands[i]}')
        try:
            action = input("\nWhat would you like to do ")
            action = int(action)
        except:
            print("\nInvalid Input ! \n")
        price_sku = []

        for i in valid_commands:
            if action == 1:
                for i,v in customer.items.items():
                    print(f'{i[-1]} : {next(iter(v))}')
                s =input("What do you like to add to cart ? ")
                customer.add(s)
                break
            elif action == 2:
                o= 0
                for i,v in customer.cart_items.items():
                    print(f'{o+1}){i}:{v}')
                    o+=1
                s =input("What item would you like to remove from cart ? ")
                customer.remove(s)
                break   
            elif action == 3:

                customer.prices()
                break
            elif action == 4:
                customer.show_cart()
                break
            elif action ==5:
                customer.check_out()
                true=False
                break
```

```python
order()
```
